```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler# for normalizing
import pickle

# Load your dataset
df = pd.read_csv("/content/us_congestion_2016_2022_sample_2m (1).csv")

# Step 1: Extract Relevant Columns
columns_needed = ['Start_Lat', 'Start_Lng', 'Congestion_Speed',
                  'Distance(mi)', 'DelayFromTypicalTraffic(mins)', 'DelayFromFreeFlowSpeed(mins)']
df = df[columns_needed]

# Convert Congestion_Speed to numeric, replacing invalid values with NaN
df['Congestion_Speed'] = pd.to_numeric(df['Congestion_Speed'], errors='coerce')

# Step 2: Define the Spatial Grid
# Determine the min and max for latitude and longitude
lat_min, lat_max = df['Start_Lat'].min(), df['Start_Lat'].max()
lng_min, lng_max = df['Start_Lng'].min(), df['Start_Lng'].max()

# Define grid resolution (e.g., 10x10 grid)
grid_size = 10 #pxl
lat_bins = np.linspace(lat_min, lat_max, grid_size + 1)#intervals frim latitude and longitude
lng_bins = np.linspace(lng_min, lng_max, grid_size + 1)

# Add grid cell information to the dataset
df['lat_bin'] = pd.cut(df['Start_Lat'], bins=lat_bins, labels=False)## divides continuous data into discrete intervals
df['lng_bin'] = pd.cut(df['Start_Lng'], bins=lng_bins, labels=False)

# Step 3: Aggregate Features into the Grid
# Initialize a 3D grid for the CNN input (grid_size x grid_size x channels)
cnn_input = np.zeros((grid_size, grid_size, 3))  # 3 channels for features

# Populate the grid
for _, row in df.iterrows():
    lat_idx = row['lat_bin']
    lng_idx = row['lng_bin']
    if not np.isnan(lat_idx) and not np.isnan(lng_idx):
        lat_idx, lng_idx = int(lat_idx), int(lng_idx)#### based on sensors we are initializing grid
        cnn_input[lat_idx, lng_idx, 0] += row['Congestion_Speed'] if not np.isnan(row['Congestion_Speed']) else 0
        cnn_input[lat_idx, lng_idx, 1] += row['Distance(mi)'] if not np.isnan(row['Distance(mi)']) else 0
        cnn_input[lat_idx, lng_idx, 2] += row['DelayFromTypicalTraffic(mins)'] if not np.isnan(row['DelayFromTypicalTraffic(mins)']) else 0

        #### based on sensors we are initializing grid

# Normalize the grid data
scaler = MinMaxScaler()
cnn_input = scaler.fit_transform(cnn_input.reshape(-1, cnn_input.shape[-1])).reshape(cnn_input.shape)

# Add batch dimension for CNN input
cnn_input = np.expand_dims(cnn_input, axis=0)  # Shape: (1, grid_size, grid_size, channels)

# Step 4: Manually Implement CNN Layers (Without TensorFlow/Keras)

# Convolution Operation
def conv2d(input, kernel):### Filter -> input image-> feature map
    kernel_height, kernel_width, input_channels, output_channels = kernel.shape
    input_height, input_width, input_channels = input.shape
    output_height = input_height - kernel_height + 1 #forml
    output_width = input_width - kernel_width + 1
    output = np.zeros((output_height, output_width, output_channels))

    for i in range(output_height):
        for j in range(output_width):
            for k in range(output_channels):
                output[i, j, k] = np.sum(input[i:i+kernel_height, j:j+kernel_width, :] * kernel[:, :, :, k])
    return output

# Max Pooling Operation
def max_pooling(input, pool_size):
    input_height, input_width, channels = input.shape
    pool_height, pool_width = pool_size
    output_height = input_height // pool_height #forml
    output_width = input_width // pool_width
    output = np.zeros((output_height, output_width, channels))

    for i in range(output_height):
        for j in range(output_width):
            for k in range(channels):
                output[i, j, k] = np.max(input[i*pool_height:(i+1)*pool_height, j*pool_width:(j+1)*pool_width, k])
    return output

# Define kernel for convolution (3x3, 3 channels to 16 channels)
kernel = np.random.randn(3, 3, 3, 16)

# Apply convolution on the grid
conv_output = conv2d(cnn_input[0], kernel)
print("Conv Output Shape:", conv_output.shape)

# Apply max-pooling (2x2)
pool_output = max_pooling(conv_output, (2, 2))
print("Pooling Output Shape:", pool_output.shape)

# Step 5: Flatten the Output and Apply a Dense Layer (Fully Connected Layer)

# Flatten the pooled output
flattened_output = pool_output.flatten()

# Define Dense Layer
def dense_layer(input, units):
    weights = np.random.randn(input.shape[0], units)
    biases = np.random.randn(units)
    return np.dot(input, weights) + biases

# Apply Dense Layer
dense_output = dense_layer(flattened_output, 64)  # 64 units in the dense layer
print("Dense Layer Output Shape:", dense_output.shape)

# Activation function: Sigmoid for output
output = 1 / (1 + np.exp(-dense_output))  # Sigmoid function
print("Spatial Features", output)

# Step 6: Save the Model (Weights and Biases)####################no
model_params = {
    'conv_kernel': kernel,
    'dense_weights': np.random.randn(flattened_output.shape[0], 64),  # Dense layer weights
    'dense_biases': np.random.randn(64)  # Dense layer biases
}

with open('cnn_model_params.pkl', 'wb') as f:
    pickle.dump(model_params, f)

# Step 7: Load the Saved Model
with open('cnn_model_params.pkl', 'rb') as f:
    loaded_params = pickle.load(f)

#print("Loaded Model Params:", loaded_params)
```

```python
import numpy as np

# Example: Spatial features (output from your Dense layer)
spatial_features = output  # Output from Dense layer
print(f"Spatial Features Shape: {spatial_features.shape}")

# Reshape spatial features for LSTM (sequence length = 10, feature size = len(spatial_features)//10)
sequence_length = 10
feature_size = len(spatial_features) // sequence_length
spatial_features = spatial_features[:sequence_length * feature_size]  # Ensure it's divisible
lstm_input = spatial_features.reshape(sequence_length, feature_size)  # Shape: (seq_len, feature_size)

print(f"LSTM Input Shape: {lstm_input.shape}")

# Step 6: Define Manual LSTM
def lstm_cell(input_seq, hidden_state, cell_state, weights):
    """
    Manually compute LSTM cell operations for a single time step.
    Args:
        input_seq (ndarray): Input at the current time step (features,).
        hidden_state (ndarray): Hidden state from the previous time step.
        cell_state (ndarray): Cell state from the previous time step.
        weights (dict): Weights and biases for LSTM gates.
    Returns:
        hidden_state, cell_state: Updated hidden and cell states.
    """
    input_size = input_seq.shape[0]
    hidden_size = hidden_state.shape[0]

    # Extract weights #each cell have 3 gates and 1 candidate cell
    Wf, Uf, bf = weights['Wf'], weights['Uf'], weights['bf']  # Forget gate
    Wi, Ui, bi = weights['Wi'], weights['Ui'], weights['bi']  # Input gate
    Wo, Uo, bo = weights['Wo'], weights['Uo'], weights['bo']  # Output gate
    Wc, Uc, bc = weights['Wc'], weights['Uc'], weights['bc']  # Candidate cell

    # Forget gate
    forget_gate = 1 / (1 + np.exp(-(np.dot(Wf, input_seq) + np.dot(Uf, hidden_state) + bf)))

    # Input gate
    input_gate = 1 / (1 + np.exp(-(np.dot(Wi, input_seq) + np.dot(Ui, hidden_state) + bi)))
    candidate_cell = np.tanh(np.dot(Wc, input_seq) + np.dot(Uc, hidden_state) + bc)

    # Update cell state
    cell_state = forget_gate * cell_state + input_gate * candidate_cell

    # Output gate
    output_gate = 1 / (1 + np.exp(-(np.dot(Wo, input_seq) + np.dot(Uo, hidden_state) + bo)))
    hidden_state = output_gate * np.tanh(cell_state)

    return hidden_state, cell_state

# Step 7: Initialize Weights and States
feature_size = lstm_input.shape[1]
hidden_size = 32  # Define LSTM hidden state size

weights = {
    'Wf': np.random.randn(hidden_size, feature_size),
    'Uf': np.random.randn(hidden_size, hidden_size),
    'bf': np.random.randn(hidden_size),
    'Wi': np.random.randn(hidden_size, feature_size),
    'Ui': np.random.randn(hidden_size, hidden_size),
    'bi': np.random.randn(hidden_size),
    'Wo': np.random.randn(hidden_size, feature_size),
    'Uo': np.random.randn(hidden_size, hidden_size),
    'bo': np.random.randn(hidden_size),
    'Wc': np.random.randn(hidden_size, feature_size),
    'Uc': np.random.randn(hidden_size, hidden_size),
    'bc': np.random.randn(hidden_size),
}

hidden_state = np.zeros(hidden_size)  # Initial hidden state
cell_state = np.zeros(hidden_size)  # Initial cell state

# Step 8: Process Input Through LSTM
for t in range(sequence_length):
    input_seq = lstm_input[t]
    hidden_state, cell_state = lstm_cell(input_seq, hidden_state, cell_state, weights)

# Final hidden state can be used as the LSTM output
print(f"Final Hidden State Shape: {hidden_state.shape}")
print("LSTM Output:", hidden_state)
```
